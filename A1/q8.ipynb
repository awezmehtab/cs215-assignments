```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('./data/Mona_Lisa.jpg')
```

```python
plt.axis('off')
plt.imshow(img)
```

```python
height = img.shape[0]
width = img.shape[1]
```

```python
def shift(img, tx):
    height, width, _ = img.shape
    if tx > 0:
        for i in range(width-1, tx-1, -1):
            img[:, i, :] = img[:, i-tx, :]
        for i in range(tx):
            img[:, i, :] = [0, 0, 0]  
    elif tx < 0:
        for i in range(0, width+tx):
            img[:, i, :] = img[:, i-tx, :]
        for i in range(width+tx,width):
            img[:, i, :] = [0, 0, 0]  
    return img
```

```python


img2 = shift(img.copy(), -10)
plt.imshow(img2)
plt.show()
```
