```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv("parkingLot.csv")
df
```

```python
print(df.info())
print(df.columns)
```

Converted to string and removedd spaces from the column name

```python
df['vehicle_no'] = df['vehicle_no '].str.upper().astype('string')
df['timestamp'] = df['timestamp           '].str.upper().astype('string')
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df = df.drop('vehicle_no ', axis=1)
df = df.drop('timestamp           ', axis=1)
df
```

```python
df.info()
```

removed spaces from the values

```python
df['vehicle_no'] = df['vehicle_no'].str.strip()
df
```

sorted by time

```python
df = df.sort_values(by='timestamp')
df
```

removed empty data

```python
df = df[df['vehicle_no'] != '']
df
```

removed that cars that appeared in the dataframe odd number of times

```python
vehicle_counts = df['vehicle_no'].value_counts()
odd_vehicles = vehicle_counts[vehicle_counts % 2 == 1].index
odd_vehicles 
```

```python
df = df[~df['vehicle_no'].isin(odd_vehicles)]
df
```

number of cars in the parking lot

```python
car_count = 0


def update_count(row):
    global car_count
    if row['camera_id'] == 1:
        car_count += 1
    elif row['camera_id'] == 2:
        car_count -= 1
    return car_count

df['cars_in_parking'] = df.apply(update_count, axis=1)

df
```

```python
df.to_csv('clean_data.csv',  index=False)
```
